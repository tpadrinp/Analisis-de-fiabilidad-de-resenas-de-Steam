Importamos librerias y definimos variables base para facilitar el entendimiento


```python
import pandas as pd

RAW_PATH = "../data/raw" # Indica el directorio donde se encuentran los datos sin transformar
PROCESSED_PATH = "../data/processed" # Indica el directorio donde se guardar√°n los datos transformados
```

Extraemos los CSV obtenidos de Kaggle

```python
# Definimos una funci√≥n para extraer datos de un archivo CSV
def extract_csv_data(data):
    path = RAW_PATH + "/" + data # Concatenamos el directorio con el nombre del archivo
    return pd.read_csv(path)

# Definimos una funci√≥n para extraer datos de un archivo JSON
def extract_json_data(data):
    path = RAW_PATH + "/" + data
    # Como el archivo JSON tiene m√∫ltiples l√≠neas, usamos lines=True
    return pd.read_json(path, lines=True)
```

Transformaci√≥n de los datos:
Al revisar los datos tras extraerlos, observamos que games_metada.json tiene la columna "tags" una lista delimitada por [], transformamos para que no aparezcan los [] y eliminamos las lineas en las que tanto "description" y "tags" est√©n vac√≠as

```python
#Definimos una funci√≥n para transformar el archivo json a un formato v√°lido
def transfomr_json(data):
    # Primero, normalizamos la columna "tags", eliminando los caracteres no deseados
    data["tags"] = data["tags"].astype(str).str.replace("[", "").str.replace("]", "").str.replace("'", "").str.replace('"', "")
    # Luego, eliminamos los registros que no tienen ni descripci√≥n ni etiquetas
    data = data[(data["description"].str.strip() != "") & (data["tags"] != "")]
    return data
```

Comprobamos que el resultado de la transformaci√≥n deseada devuelve los datos tal y como queremos

```python
df = extract_json_data("games_metadata.json")
transfomr_json(df)
```
